```python
conn = sqlite3.connect("../../db.sqlite3")
cur = conn.cursor()
```

```python
query = cur.execute("""
    SELECT 
    url,
    description,
    articleBody,
    score
    FROM news
""")
```

```python
conn.close()
```

```python
data = query.fetchall()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, 
                  columns=['url', 
                           'desc', 
                           'body',
                           'score'
                          ])
print(df.shape)
df_no_desc = df[df["desc"] == ""]
df_with_desc = df[df["desc"] != ""]
df_with_desc["score"] = df_with_desc["score"] + 1
print(df_with_desc.shape)
df_with_desc.head()
```

Unique values by column

```python
df_with_desc["url"].nunique(), df_with_desc["desc"].nunique(), df_with_desc["body"].nunique()
```

Extract first occurence from each 'desc' column group

```python
df_first_from_groups = df_with_desc.groupby("desc").first().reset_index()
df_first_from_groups
```

Extract random occurence from each 'desc' column group

```python
index_one_random_from_groups = df_with_desc.groupby("desc") \
                                           .sample(n=1, weights="score") \
                                           .index
not_index_one_random_from_groups = df.index.difference(index_one_random_from_groups)
index_one_random_from_groups, not_index_one_random_from_groups
```

```python
ex_desc = '"Estoy convencido de que el pueblo espa√±ol ante momentos dif√≠ciles saca su mejor cara", ha dicho Ignacio Garriga'
df_example = df_with_desc[df_with_desc["desc"] == ex_desc]
df_example.sample(n=1, weights='score')["url"], \
df_example.sample(n=1, weights='score')["url"], \
df_example.sample(n=1, weights='score')["url"], \
df_example.sample(n=1, weights='score')["url"]
```

```python
urls_preprocessed = df.iloc[index_one_random_from_groups]["url"].tolist()
urls_preprocessed = tuple((x, ) for x in urls_preprocessed)
urls_only_updated = df.iloc[not_index_one_random_from_groups]["url"].tolist()
urls_only_updated = tuple((x, ) for x in urls_only_updated)
urls_preprocessed, urls_only_updated
```

```python
conn = sqlite3.connect("../../db.sqlite3")
cur = conn.cursor()
```

```python
cur.executemany("""
    UPDATE  
        news
    SET
        preprocessed = True,
        updateDate = DATETIME('now', 'localtime', 'utc')
    WHERE
        url = ?
""", urls_preprocessed)
conn.commit()
cur.executemany("""
    UPDATE  
        news
    SET
        updateDate = DATETIME('now', 'localtime', 'utc')
    WHERE
        url = ?
""", urls_only_updated)
```
